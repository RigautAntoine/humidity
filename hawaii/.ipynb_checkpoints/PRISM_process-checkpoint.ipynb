```python
import pandas as pd
import numpy as np
import os
import xarray as xr
from cartopy import crs
import panel as pn
import bokeh as bk
import hvplot.pandas
import hvplot.xarray
import holoviews as hv

##### EDIT #####
import datashader as dsh
from holoviews.operation.datashader import regrid
import geoviews as gv
#gv.extension('bokeh')

# Import the functions needed
from functions import *
compute_RH(20, 15, kelvin=False)
```

```python
dem = xr.open_dataset('data/hi_dem_15s.asc')#['band_data'].mean(dim='band')
dem['pressure'] =  pressure(dem['band_data'])
dem['pressure'].plot()
```

```python
# TDMEAN unit is Celsius x 100
# Divide by 100 to get back to expectaction
#ds = xr.open_dataset('data/PRISM_tdmean_hi_30yr_normal_80mM1_07_asc.asc')['band_data'].mean(dim='band') / 100
```

```python
months = np.arange(1,13)
ds = (
    xr.concat([
        xr.open_dataset('data/PRISM_tdmean_hi_30yr_normal_80mM1_{:02d}_asc.asc'.format(i)) for i in months],
        dim = pd.Index(months, name='month'))
    .sel(band=1).drop('band'))/ 100
```

```python
ds
```

```python
# Compute vapor pressure from the Celsius temperatures
ds['vapor_pressure'] = vapor_pressure(ds['band_data'], kelvin=False)
```

```python
# Add atmospheric pressure from orology data
ds['atmospheric_pressure'] = dem['pressure']
```

```python
# Need to drop band
ds['humidity'] = compute_SH(ds['vapor_pressure'], ds['atmospheric_pressure']) #(0.622 * ds['vapor_pressure']) / (ds['atmospheric_pressure'] - 0.378 * ds['vapor_pressure']) * 1000
```

```python
# Write to the app's data folder
ds.to_netcdf("../app/data/hawaii_30y.nc")
```

```python
## Graphic
```

```python
# Select
ds2 = ds['humidity'].sel(band = 1).rio.write_crs(4326)
```

```python
ds2
```

```python
custom_hover = bk.models.HoverTool(tooltips=[('Specific Humidity', '@image{0} g/kg')])

#### EDIT START
gvds = gv.Dataset(ds2)
images = gvds.to(gv.Image, ['x', 'y'], dynamic=True)

base_map = hv.element.tiles.OSM()

plot_map = (base_map * regrid(images).opts(
    cmap='RdBu',
    clim=(0,20),
    alpha=0.7,
    #geo=True,
    colorbar=True,
    tools=[custom_hover],
    width=900, 
    height=500))#.opts(projection=crs.PlateCarree()) #* gv.feature.coastline
```
